```python
#standardize header names. 

cols = [] 

for i in range(len(costumer_analysis.columns)):
    cols.append(costumer_analysis.columns[i].lower().replace(' ','_'))
costumer_analysis.columns = cols
costumer_analysis
```

```python
#check for null values and show the table 
condition = costumer_analysis.isnull().any(axis=1)
costumer_analysis[condition]
```

```python
costumer_analysis.isna().sum() / len(costumer_analysis) # check the percentage of the null values, the vehicle_variable has 50% of null values 
```

```python
null_data = costumer_analysis.isnull()[['state', 'response', 'months_since_last_claim', 'number_of_open_complaints', 'vehicle_class', 'vehicle_size']]
costumer_analysis[null_data] #eliminating all the null values from the the specified variables, first I check if they had ima
```

```python
drop_na = costumer_analysis.dropna(subset = null_data.columns) #
costumer_analysis = drop_na
costumer_analysis
```

```python
costumer_analysis['income'].sum() #if we drop the vehicle_type the income lowers a lot 
```

```python
costumer_analysis
```

```python
costumer_analysis['vehicle_type'] = costumer_analysis['vehicle_type'].astype(str)
costumer_analysis.dtypes

costumer_analysis['vehicle_type'] = pd.to_numeric(costumer_analysis['vehicle_type'], errors='coerce')
costumer_analysis.dtypes
```

```python
costumer_analysis['vehicle_type']=costumer_analysis['vehicle_type'].fillna(costumer_analysis['vehicle_type'].mean())
costumer_analysis.isna().sum()
```

```python
#ther's still 9134 null values that I cant eliminate 

vehicle_type_null = costumer_analysis.isnull()['vehicle_type']
costumer_analysis[vehicle_type_null]

drop_vehicle_type = costumer_analysis.dropna(subset = vehicle_type_null.columns)
costumer_analysis = drop_vehicle_type
costumer_analysis.isna().sum()

```
