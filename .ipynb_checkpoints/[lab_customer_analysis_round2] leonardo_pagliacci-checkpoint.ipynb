```python
# Datetime format - Extract the months from the dataset and store in a separate column. 
# Then filter the data to show only the information for the first quarter , ie. January, February and March.
# _Hint_: If data from March does not exist, consider only January and February.

# Converting to date-time dtype
customer_no_na['effective_to_date'] = pd.to_datetime(customer_no_na['effective_to_date'])


customer_no_na['month'] = customer_no_na['effective_to_date'].dt.month


customer_no_na

```

```python

condition = (customer_no_na['month'] >= 1) | (customer_no_na['month'] >= 2) | (customer_no_na['month'] <= 3) 

customer_Q1 = customer_no_na[condition]

customer_Q1
```

```python
# Now i convert the month number in 'number' column in the actual month names

month_names = []

month_name_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Iterate through the month column and convert to month names
for i in customer_Q1['month']:
    month_name = month_name_list[i - 1]
    month_names.append(month_name)

# Create a new column in the DataFrame with the month names
customer_Q1['month'] = month_names

customer_Q1
```
