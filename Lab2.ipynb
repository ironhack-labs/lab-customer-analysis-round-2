```python
def clean_data(data1):
    import pandas as pd
    import numpy as np

    data1 = pd.read_csv(data1)
    print("Data shape:", data1.shape)

    data1.columns = data1.columns.str.lower().str.replace(' ', '')
    print("Standardized data columns: ", data1.columns)

    numerical_cols = data1.select_dtypes(include=['int', 'float']).columns
    print("Numerical Columns: ", numerical_cols)

    categorical_cols = data1.select_dtypes(include=['object']).columns
    print("Categorical Columns: ", categorical_cols)

    # clean data
    nulls_df = pd.DataFrame(data1.isna().sum() / len(data1) * 100).reset_index()  # formatar em tabela
    nulls_df.columns = ['column_name', 'percent_nulls']

    columns_to_drop = nulls_df[nulls_df['percent_nulls'] >= 50]['column_name'].values
    data1 = data1.drop(columns_to_drop, axis=1)

    data1 = data1.drop_duplicates()

    data1 = data1.dropna(axis=0)

    import warnings

    warnings.filterwarnings("ignore")

    data1['effectivetodate'] = pd.to_datetime(data1['effectivetodate'], errors='coerce')
    data1['month'] = data1['effectivetodate'].dt.month

    data1 = data1[data1['month'].isin([1, 2, 3])]

    print("Final data filtered is: ")

    return pd.DataFrame(data1)


# test
clean_data("marketing_customer_analysis.csv")
```
