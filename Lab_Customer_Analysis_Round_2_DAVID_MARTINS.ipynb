```python
# CHECK AND DEAL WITH NaN VALUES

# Lets now work with null values

customer_data.isnull()

customer_data.isna()

customer_data.isna().sum() 

customer_data.isna().sum().sum()

customer_data.dropna()

customer_data.isna().sum()

customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].fillna(customer_data['number_of_open_complaints'].mean())
customer_data.isna().sum()


```

```python
# DATETIME FORMAT- EXTRACT THE MONTHS FROM THE DATASET AND STORE IN A SEPARATE COLUMN. THEN FILTER THE DATA TO SHOW ONLY THE REST OF THE QUARTER.
# ie. January, February, and March. Hint: IF DATA FROM MARCH DOES NOT EXIST, CONSIDER ONLY JANUARY AND FEBRUARY.



customer_data['effective_to_date'] = pd.to_datetime(customer_data['effective_to_date'])

customer_data['month'] = customer_data['effective_to_date'].dt.month

customer_data    # now i have i new column with months

# Filterinf the January, February, March month 


filtered_rows = customer_data[customer_data['month'] < 4]

print(filtered_rows)

def filter_rows_by_month(dataframe, condition):
    filtered_rows = dataframe[condition]
    return filtered_rows


filtered_data = filter_rows_by_month(customer_data, customer_data['month'] < 4)
print(filtered_data)

customer_data
```
