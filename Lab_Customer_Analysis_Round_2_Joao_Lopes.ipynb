```python
# Deal with NaN values

initial_income = df["income"].sum()
print("Initial total income:", initial_income)

df["months_since_last_claim"].fillna(int(df["months_since_last_claim"].mean()), inplace = True)
df["number_of_open_complaints"].fillna(int(df["number_of_open_complaints"].mean()), inplace = True)
df["vehicle_type"].fillna("A", inplace = True)
df = df.dropna()
df = df.reset_index(drop = True)

final_income = df["income"].sum()
print("Final total income:", final_income)
print("Percentage of loss income:", int((1-final_income/initial_income)*100), "%")
df
```
