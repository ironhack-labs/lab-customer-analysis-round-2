### Importing Libraries and Opening the file

```python
import pandas as pd
import numpy as np
import datetime as dt
df = pd.read_csv('marketing_customer_analysis.csv')
```

### 1. Show the dataframe shape.

```python
df.shape
```

Our dataset has 10910 unique data information and 26 columns. Let's see how it looks:

```python
df
```

### 2. Standardize header names.

Previously, we opened the dataset and we can see that there are some things to correct in their columns.
As we did in the previous lab and as a must, we have to first replace all column names to lower capital letters and also replace any `space` for `_`

```python
df.columns = cc = [column.lower().replace(' ', '_') for column in df]
```

Now all columns have the standardized format. We detect that in one column (`unnamed:_0`) there is also a number at the end.
Also, we cannot see all columns so we use another `method` from pandas to see only the columns:

```python
df.columns
```

The other they just look just fine, so we will just change the first one, deleting the ":_0" for readability,

```python
df.columns = cc = [column.replace(':_0', '') for column in df]
df.columns
```

### 3. Which columns are numerical?

9 of our columns are numerical.

### 4. Which columns are categorical?

```python
df.select_dtypes(exclude=["float", "int"])
```

And 17 of them are categorical.

### 5. Check and deal with `NaN` values.
We spot that before, now we have to deal with all `Nan` values.
Since we have them in both `numerical` and `categorical` columns, our approach will be different:

```python
# For the numerical columns, we will replace all NaN values with 0 with the .fillna() method from pandas.
df.select_dtypes(include=["float", "int"]).fillna(0)
df.select_dtypes(exclude=["float", "int"]).fillna('')
```

```python
# We print the result to see the changes in both datasets
df
```

### 6. Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. _Hint_: If data from March does not exist, consider only January and February.

```python
#  Extract the months from the dataset and store in a separate column
df['month'] = pd.DatetimeIndex(df['effective_to_date']).month

# How many unique months do we have?
df["month"].value_counts()
```

```python
# In our dataframe we have 5818 entries for January, 5092 for February. We will consider only January and February for our filter:
df = df[df['month'].isin([1,2])]
df
```
