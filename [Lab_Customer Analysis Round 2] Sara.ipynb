```python
#BONUS 

ct_data = pd.read_csv('marketing_customer_analysis.csv')
ct_data

def ct_data_done(ct_data):
    ct_data = ct_data.copy()
    ct_data.columns = ct_data.columns.str.strip().str.lower().str.replace(' ', '_')
    numerical_columns = ct_data.select_dtypes(include=[np.number]).columns
    categorical_columns = ct_data.select_dtypes(exclude=[np.number]).columns
    nan_counts = ct_data.isna().sum()
    ct_data[numerical_columns] = ct_data[numerical_columns].fillna(ct_data[numerical_columns].mean())
    ct_data = ct_data.dropna()
    ct_data.loc[:, 'date'] = pd.to_datetime(ct_data['effective_to_date'])
    ct_data.loc[:, 'month'] = ct_data['date'].dt.month
    ct_data = ct_data[ct_data['month'] <= 3]
    return ct_data
ct_data_done
ct_data = ct_data_done(ct_data)
ct_data
```
