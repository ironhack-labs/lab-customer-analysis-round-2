```python
# Checking for NaNs

sum_na = df.isna().sum()
print (sum_na)


'''Interpretation:
- The columns 'state', 'response', 'months_since_last_claim', 'number_of_open_complaints', 'vehicle_class', 'vehicle_size',
and 'vehicle_type' have NaN values
'''

'''
How I would deal with these values:
1. Calculate the % of the NaA values for the specific column
2.1 If nummeric column, check mean and sd to see if filling the data with the mean would be appropriate
2.2 If categorical, decide what information are given in this column and if they are necessary for the actual analysis of the 
    data. If not, consider dropping or leaving the values --> for decison include % of NaN values in this particular column
    
'''
```

```python
# Dealing with these NaN values
# Calculating % of NaN in df

nan_percent = df.isna().sum() / len(df)*100
print(nan_percent)

'''
Interpretation:
- Every column with NaN values, except for 'vehicle_type' has ~ 6 % of NaA values. 'vehicle_type' has >50 %
'''

# Calulating the basic descriptive statistics of the numerical columns
num_stat = df[['months_since_last_claim', 'number_of_open_complaints']].describe()
print (num_stat)

'''
Interpretation:
- As for both cloumns sd is relatively high, filling the NaN values with the mean of the df could lead to a biased 
    interpreatation of the data; (obously depending on what factors/values will be analysed)
- Having said this, I would keept the NaN values, as mathematical calculations are still possible with them in the df.
'''


'''
Categorical columns:
Here it depends on the target of the analysis. As for all categorical columns holding NaNs, except for 'vehicle_type', 6 %
are just a small percentage of the whole data and these rows could be dropped if necessary; for example if they don't have an 
influence on the 'income'. Concerning 'vehicle_type' it should be assessed if this coloumn is necessary for the analysis, as it
is present in less than 50% of the samples and the only value in it is 'A'. I think the whole column could be dropped without any
'harm'.

'''


```

```python
# Decision for the cateorgical data
df['vehicle_type'].unique()
```
