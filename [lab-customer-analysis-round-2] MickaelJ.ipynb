```python
# Importe Date and time librairie/ create a new column in data frame (month)/ use panda to date time to excract date from 'effective date' columns/ print data frame 
# .dt stands for "datetime accessor," and it's a special attribute that allows you to access various components of datetime objects in a Series or DataFrame
# convert Data in perido Month to be able to organise per month the data
import datetime
import pandas as pd

mark_cust_A['Date'] = pd.to_datetime(mark_cust_A['effective_to_date'])

mark_cust_A['Date'] = mark_cust_A['Date'].dt.to_period('M')

# first condition creating a new data Frame for January and adding a column date and month 
condition1 = mark_cust_A['Date'] == '2011-01'
data_for_jan =mark_cust_A[condition1]
data_for_jan['month'] = 'Jan'



## second condition creating a new data Frame for February and adding a column date and month 
condition2 = mark_cust_A['Date'] == '2011-02'
data_for_fev = mark_cust_A[condition2]
data_for_fev['month'] = 'Fev'


## third condition creating a new data Frame for March and adding a column date and month 
condition3 = mark_cust_A['Date'] == '2011-03'
data_for_march = mark_cust_A[condition3]
data_for_march['month'] = 'March'


# concatenate the dataframe for every month to get back to Data frame original with month ordered

mark_cust_df = pd.concat([data_for_jan,data_for_fev,data_for_march], axis = 0)
mark_cust_df
```

```python
# import Data and read excel file with special path as the excel is in my Google drive , import panda library,

import pandas as pd
url = 'https://docs.google.com/spreadsheets/d/1a9g5MfgAJ17HiYYKsjBi2yaO_HMGLIAX/edit?usp=sharing&ouid=108054823958769193495&rtpof=true&sd=true'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
mark_cust_A = pd.read_excel(path)
# fonction processing all the steps to make the data clean for our file
def Data_quarter_Y(mark_cust_A):
    # creation of a list to store columns name after lower and removing spaces
    cols_st = []
    # 
    for i in range(len(mark_cust_A.columns)):
        cols_st.append(mark_cust_A.columns[i].lower().replace(" ","_"))
    mark_cust_A.columns = cols_st
    
    mark_cust_A = mark_cust_A.fillna({'response':'No','number_of_open_complaints': 0,'vehicle_type':'Other vehicule'})
    mark_cust_A = mark_cust_A.drop("unnamed:_0", axis=1)
    mark_cust_A['date'] = pd.to_datetime(mark_cust_A['effective_to_date'])
    mark_cust_A['month'] = mark_cust_A['date'].dt.to_period('M')
    
    condition1 = (mark_cust_A['date'] == '2011-01')
    condition2  = (mark_cust_A['date'] == '2011-02')
    condition3 = (mark_cust_A['date'] == '2011-03')
    
    data_for_jan = mark_cust_A[condition1].copy()
    data_for_jan['month'] = 'Jan'

    data_for_fev = mark_cust_A[condition2].copy()
    data_for_fev['month'] = 'Feb'

    data_for_march = mark_cust_A[condition3].copy()
    data_for_march['month'] = 'March'
      
    return pd.concat([data_for_jan, data_for_fev, data_for_march], axis=0)   

mark_cust_A = Data_quarter_Y(mark_cust_A)

mark_cust_A
```
