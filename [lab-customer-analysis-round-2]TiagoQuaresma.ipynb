```python
condition=data.isnull().any(axis=1)
data[condition][['state','response','months_since_last_claim','number_of_open_complaints','vehicle_class','vehicle_size']]

#Since the percentage is singnificante I will find a way to fill them.
```

```python

#for state
#data['state'] = data['state'].fillna(data['state'].mode())
#data['state'] = data['state'].fillna(data['state'].mode)
data['state'] = data['state'].fillna(data['state'].mode().iloc[0]) 
#for response
data['response'] = data['response'].fillna(data['response'].mode)
#for months_since_last_claim 
# Since the mean is sensitive to outliers, I will use the median to keep the dataset consistent and unbias
data['months_since_last_claim'] = data['months_since_last_claim'].fillna(data['months_since_last_claim'].median())
#for number_of_open_complaints
data['number_of_open_complaints'] = data['number_of_open_complaints'].fillna(data['number_of_open_complaints'].mode().iloc[0])
#for vehicle_class
data['vehicle_class'] = data['vehicle_class'].fillna(data['vehicle_class'].mode().iloc[0])
#for vehicle_size	
data['vehicle_size'] = data['vehicle_size'].fillna(data['vehicle_size'].mode().iloc[0])
```

### 6 - Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
data['effective_to_date_og']= pd.to_datetime(data['effective_to_date'])
data
```

```python
data['month']= data['effective_to_date_og'].dt.month
data
```

```python
#assuming there is no -1 or something
first_quarter= data[data['month']<=3]
first_quarter
```

```python
def data_tranformation(df, coluna_date):
    df[coluna_date]= pd.to_datetime(df[coluna_date])
    
    df['quarter']= df[coluna_date].dt.quarter
    first_quarter= df[df['quarter']==1]
    
    return first_quarter

```

```python
data_tranformation(data, 'effective_to_date')
```
