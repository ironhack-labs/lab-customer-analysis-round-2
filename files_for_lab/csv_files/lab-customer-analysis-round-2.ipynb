```python
def transform_data(data):
    data['effective to date'] = pd.to_datetime(data['effective to date'],errors='coerce')
    data['month'] = pd.DatetimeIndex(data['effective to date']).month
    data = data[(data['month'] > 0) & (data['month'] < 4)]
    return data
```
