```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
customer_data = pd.read_csv(r'files_for_lab\csv_files\marketing_customer_analysis.csv')
customer_data
```

```python
# Standadize header names

customer_data = customer_data.rename(columns={'EmploymentStatus' : 'EMPLOYMENT_STATUS'})

cols = []
for i in range(len(customer_data.columns)):
    cols.append(customer_data.columns[i].lower().replace(' ', '_'))

customer_data.columns = cols
customer_data
```

```python
# Wich columns are numerical?

num_cols = customer_data.describe(include=['float','int'])
num_cols
```

```python
# Which columns are categorical?

categorical_cols = customer_data.describe(include=['object'])
categorical_cols
```

```python
# Check and deal with NaN values

customer_data['number_of_open_complaints'].unique()
customer_data['number_of_open_complaints'].count()
```

```python
customer_data.dropna()
customer_data['number_of_open_complaints'] = customer_data['number_of_open_complaints'].fillna(customer_data['number_of_open_complaints'].mean())
customer_data.isna().sum()
```

```python
customer_data['effective_to_date'] = pd.to_datetime(customer_data['effective_to_date'])
customer_data['month'] = customer_data['effective_to_date'].dt.month
customer_data
```

```python
# Filter the data to show only the information for the first quarter

filtered_rows = customer_data[customer_data['month'] < 4]
filtered_rows

def filter_by_month(dataframe, condition):
    filtered_rows = dataframe[condition]
    return filtered_rows


filtered_data = filter_by_month(customer_data, customer_data['month'] < 4)
print(filtered_data)

customer_data
```
