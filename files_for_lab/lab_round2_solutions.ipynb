**5. Check and deal with NaN values.**

```python
df.isnull().sum()
```

```python
cat = []
con = []
for i in df.columns:
    if(df[i].dtypes =='object'):
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
for i in cat:  
    x = df[i].mode()[0]  
    df[i]= df[i].fillna(x)
for i in con:
    x = df[i].mean()
    df[i]= df[i].fillna(x)
```
